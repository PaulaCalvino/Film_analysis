```python
import numpy as np
import pandas as pd
from time import sleep

import requests

from pprint import pprint
```

```python

BASE_ID = "appI9C1deMEUemvfN"

TABLE_ID = "tblA9QIFhXjh5Mjui"

airtable_base_url = "https://api.airtable.com/v0"

headers = {"Authorization" : f"Bearer {TOKEN}",
           "Content-Type"  : "application/json"}
```

```python
# Endpoint

endpoint = f"{airtable_base_url}/{BASE_ID}/{TABLE_ID}"

print(endpoint)
```

```python
# GET Records
#Obtenemos los datos de la tabla de películas subida a Airtable

params = {"offset" : None}

df = pd.DataFrame()

while params.get("offset") != None or df.shape[0] == 0:
    
    response = requests.get(url = endpoint, headers = headers, params = params)
    
    print(response.url)
    
    print(f"response: {response.status_code}")
    
    params["offset"] = response.json().get("offset")
    
    print(params.get("offset"))
    
    df = pd.concat([df, pd.json_normalize(response.json()["records"])], ignore_index = True)
    
    sleep(0.5)
    
df.shape[0]
```
